```r
install.packages("tidyverse")
library(ggplot2)
air<-airquality
head(air)
```

ggplot(data = … , mapping = aes(...)) + geom()



```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind)) + geom_point()
```

```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind, color=Month)) + geom_point(alpha = .5, color = "purple")
```

What if we change where the color is? Put it in geom_point()

geom_point(aes(color=Month))

```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind, color=Month)) + geom_point() + geom_smooth(method = "lm")
```

What if we want linear models for each month? Can play around and do shape too

```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind, color=Month, group=Month)) + geom_point() + geom_smooth(method = "lm", se=FALSE) # se from CI
```

```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind, color=Month)) + geom_point() + labs(
    title = "Average Wind Speed vs. Ozone Mean",
    x = "Ozone Mean", y = "Average Wind Speed",
    color = "Month"
  )
```

```r
ggplot(data = air, mapping = aes(x = Temp)) + geom_histogram(binwidth = 10, fill="sky blue")
```

```r
ggplot(data = iris, mapping = aes(x = Species, fill=Species)) + geom_bar(position='fill')
```

```r
ggplot(data = air, mapping = aes(x = Ozone, y = Wind)) +
  geom_point(aes(color = Temp)) +
  facet_wrap(~Month)
```

```r
ggplot(data = iris, mapping = aes(x = Sepal.Length, y = Sepal.Width)) +
  geom_smooth(aes(linetype = Species)) #color = Species
```

dplyr - makes data manipulation easier



```r
library(dplyr)

ggplot(data = airquality, mapping = aes(x = Ozone, y = Wind)) +
  geom_point() +
  geom_point(data = filter(airquality, Temp > 90),
             shape = "circle open", size = 3, color = "red") +
  geom_point(data = filter(airquality, Temp > 90),
             color = "red")
```

```r
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```r
sns.barplot(data=tips, x="day", y="total_bill")

# Using matplotlib to add labels
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```r
sns.barplot(data=tips, x="day", y="total_bill", width = 0.1)

plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```r
sns.barplot(data=tips, x="day", y="total_bill", width = 0.1, palette = 'pastel')

plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```r
sns.lmplot(data=tips, x="total_bill", y="tip")

plt.title("Total Bill vs Tips")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()
```
